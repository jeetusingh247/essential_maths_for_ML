### Day 01 : Vectors In Machine Learning

```python
import numpy as np
v=np.array([1,-1,2])
w=np.array([2,5,2])

print(v+w)
```

```python
print(v-w)
```

##### Scalar Multiplication With Vector Results a vector

```python
print(3*v)
```

##### Finding Length of a Vector

```python
print(np.linalg.norm(v))
```

#### Dot Product of 2 Vectors

```python
s=np.dot(v,w)
print(s)
```
