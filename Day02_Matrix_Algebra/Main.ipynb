#### Day 02 : Basics of Matrix Algebra

```python
import numpy as np

# 3 x 2 matrix formation
P = np.array([[1,7],[2,1],[3,2]])
print(P)
```

```python
Q = np.array([[1,1], [1,-1], [2,3]])
print(Q)
```

```python
C1 = np.add(P,Q) # adding two matrices of same size
print(C1)
```

```python
C2 = np.subtract(P,Q) # subtracting two matrices of same size
print(C2)
```

```python
R = np.array([[3,2,1],[1,2,5]])
print(R)
```

```python
C3 = np.dot(P,R)
print(C3) # matrix multiplication resulting in 3 x 3 matrix formation
```

#### Finding the determinant of matrix

```python
det = np.linalg.det(C3)
print(det)
```

```python
S2 = np.linalg.inv(C3)
print(S2)
```

```python
M=np.array([[1,-1],[2,3]])
print(np.linalg.inv(M))
```
